```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10,3))

cmap = plt.get_cmap('viridis') 

n_task = 10
t0 = 100

xs = np.arange(1100,)

plt.subplot(1,2,1)

for i in range(1, n_task+1):
    ys = (xs < (i-1) * t0) * 0 + (xs > i * t0) + (xs <= i * t0) * (xs >= (i-1) * t0) * (xs - (i-1) * t0)/t0
    plt.plot(xs, ys, color=cmap((i-1)/(n_task-1)))
    
#plt.legend(np.arange(n_task)+1)
    
plt.xlabel('steps', fontsize=15)
plt.ylabel('skill level', fontsize=15)
plt.title('Serial model', fontsize=20)


plt.subplot(1,2,2)

for i in range(1, n_task+1):
    ys = 1 - np.exp(-i**(-1)*xs/30)
    plt.plot(xs, ys, color=cmap((i-1)/(n_task-1)))
    
#plt.legend(np.arange(n_task)+1)
    
plt.xlabel('steps', fontsize=15)
plt.ylabel('skill level', fontsize=15)
plt.title('Parallel model', fontsize=20)
plt.savefig('./plot/domino_quanta_illus.pdf', bbox_inches='tight')
```
