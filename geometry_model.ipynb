```python
import torch
import numpy as np
import copy as copy

seed = 0
torch.manual_seed(seed)
np.random.seed(seed)

############# Define task vectors ############
n_dim = 1000 # dimension (number of model parameters)
n_task = 10 # number of tasks

# task frequency p_k ~ k^(-alpha)
alpha = 2 
prob = 1/(torch.arange(n_task) + 1)**(alpha)
prob /= torch.sum(prob)

task_ortho = True # If true, task vectors are orthogonalized
loss_fn_name = 'mse' # 'mse' (mean squared error) or 'ce' (cross entropy)
    
if task_ortho:
    assert n_dim >= n_task, 'task vectors cannot be made orthogonal when n_dim < n_task!'
    gaus = torch.randn(n_dim, n_dim)
    u, s, v = torch.svd(gaus) # orthogonalize task vectors
    task_vectors = u[:n_task]
else:
    u = torch.randn(n_task, n_dim)
    u = u/torch.norm(u, dim=1, keepdim=True)
    task_vectors = u
    
############# Define the optimizer ############
noise_scale = 0. # manually add Gaussian noise to gradients
batch_size = -1 # batch_size = -1 means loss is computed using the true probability instead of the empirical one
learning_rate = 3e-4
n_steps = 1000

x = torch.nn.Parameter(torch.zeros(n_dim,))
optimizer = torch.optim.Adam([x], lr=learning_rate, betas=(0.0,0.0))

############# Training ############
task_abilities = []
losses = []
losses_skill = []
n_uses = []

for i in range(n_steps):

    optimizer.zero_grad()

    # project to get skill levels
    task_ability = task_vectors @ x
    if loss_fn_name == 'ce':
        task_ability = torch.sigmoid(task_ability)
    
    
    if batch_size > 0:
        # sample a batch
        id = np.random.choice(n_task, batch_size, p=prob.detach().numpy())
        task_vectors_batch = task_vectors[id]
        task_ability_batch = task_ability[id]
        if loss_fn_name == 'mse':
            loss = torch.mean((1-task_ability_batch)**2)
        else:
            loss = torch.mean(-torch.log(task_ability_batch))
    else:
        if loss_fn_name == 'mse':
            loss = torch.mean((1-task_ability)**2*prob)
        else:
            loss = torch.mean(-torch.log(task_ability)*prob)

    loss.backward()
    
    # manually add noise
    x.grad.data += noise_scale * torch.randn(n_dim,)
    
    # collect statistics
    task_abilities.append(task_ability.detach().numpy())
    losses.append(loss.item())
    losses_skill.append((1 - task_ability.detach().numpy()) ** 2)

    optimizer.step()

task_abilities = np.array(task_abilities)
losses = np.array(losses)
n_uses = np.array(n_uses)
losses_skill = np.array(losses_skill)
```

```python
import matplotlib.pyplot as plt


cmap = plt.get_cmap('viridis') 

for i in range(n_task):
    plt.plot(task_abilities[:,i], color=cmap(i/(n_task-1)))

plt.xlabel('steps')
plt.ylabel('skill level')
```
